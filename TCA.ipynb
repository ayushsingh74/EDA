```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0") #replacing blank with 0
df["TotalCharges"] = df["TotalCharges"].astype("float") # change the data type from object to float
```

```python
def conv(value):      #converted 0 and 1 values of senior citizen to yes or no
    if value == 1:
        return "yes"
    else:
        return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
df.head(30)
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = df['Churn'],data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
plt.figure(figsize = (4,4))
gb = df.groupby('Churn').agg({"Churn":"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customers", fontsize = 10)
plt.show() #from the given pie chart we can conclude that 26.54% of our customers have churned out
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = df["gender"], data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (4,4))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.show()
```

```python
# Calculate the counts
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()
# Calculate percentages
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

# Plot
fig, ax = plt.subplots(figsize=(4, 4))
percentages.plot(kind='bar', stacked=True, ax=ax, cmap='coolwarm')
plt.title('Churn by Senior Citizen (as % of total)')
plt.ylabel('% of Total')
plt.xlabel('Senior Citizen')

# Add percentage labels on the bars
for container in ax.containers:
    labels = [f'{v:.1f}%' if v > 0 else '' for v in container.datavalues]
    ax.bar_label(container, labels=labels, label_type='center')

plt.legend(title='Churn')
plt.show()  #comparative a greater percent of senior citizen category have churned out
```

```python
# List of categorical columns to plot
cols = ['PhoneService', 'MultipleLines', 'InternetService', 
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set figure size and layout
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))  # 3x3 grid
fig.suptitle("Countplots for Different Services", fontsize=16)

# Flatten axes array for easy iteration
axes = axes.flatten()

# Loop through columns and create countplots
for i, col in enumerate(cols):
    sns.countplot(x=df[col], ax=axes[i], palette="coolwarm", hue  = df["Churn"])
    axes[i].set_title(f"Distribution of {col}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels if needed

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.96])  # To avoid overlapping with suptitle
plt.show()
```

#Across most service categories, non-churned customers (blue) generally outnumber churned ones (orange), though specific services like OnlineSecurity, TechSupport, and StreamingTV show a more balanced distribution, indicating potential churn drivers. Notably, customers without PhoneService or InternetService appear to have lower churn rates, while those with Fiber optic internet exhibit relatively higher churn. These insights suggest that service type and availability might significantly impact customer retention.
