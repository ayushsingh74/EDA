```python
sns.set_style("whitegrid")  # set the background style white

plt.figure(figsize = (12,6))
ev_adoption_by_year = df['Model Year'].value_counts().sort_index()
sns.barplot(x=ev_adoption_by_year.index, y=ev_adoption_by_year.values, palette="viridis")
plt.title('EV Adoption Over Time')
plt.xlabel('Model Year')
plt.ylabel('Number of Vehicles Registered')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
ev_county_distribution = df['Country'].value_counts()   # Count EVs by Country
top_countries = ev_county_distribution.head(3).index     # Top 3 countries
top_countries_data = df[df['Country'].isin(top_countries)]   # Filter the Original Data
ev_city_distribution_top_countries = top_countries_data.groupby(['Country', 'City']).size().sort_values(ascending=False).reset_index(name='Number of Vehicles')   # Count EVs by City Within Each County
top_cities = ev_city_distribution_top_countries.head(10)   # Get Top 10 City-Country Combinations
plt.figure(figsize=(12, 8))
sns.barplot(x='Number of Vehicles', y='City', hue='Country', data=top_cities, palette="magma")
plt.title('Top Cities in Top Counties by EV Registrations')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('City')
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

```python
# Count how many vehicles are registered for each electric vehicle type
ev_type_distribution = df['Electric Vehicle Type'].value_counts()
# Create a new figure for plotting with a specific size (width=10, height=6 inches)
plt.figure(figsize=(8, 4))
# Create a horizontal bar plot to show the number of vehicles per EV type
sns.barplot(
    x=ev_type_distribution.values,   # Bar lengths (number of vehicles)
    y=ev_type_distribution.index,    # Bar labels (EV types)
    palette="rocket"                 # Use the "rocket" color palette for styling
)
plt.title('Distribution of Electric Vehicle Types')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Electric Vehicle Type')
plt.tight_layout()
plt.show()
```

```python
# Count how many vehicles are registered for each 'Make' and get the top 10
ev_make_distribution = df['Make'].value_counts().head(10)
# Create a new plot with width=12 inches and height=6 inches
plt.figure(figsize=(12, 6))
# Draw a horizontal bar plot for the top 10 EV makes
sns.barplot(
    x=ev_make_distribution.values,  # Number of vehicles (length of bars)
    y=ev_make_distribution.index,   # Manufacturer names (labels for bars)
    palette="cubehelix"             # Use the "cubehelix" color palette
)
plt.title('Top 10 Popular EV Makes')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Make')
plt.tight_layout()
plt.show()
```

```python
# selecting the top 3 manufacturers based on the number of vehicles registered
top_3_makes = ev_make_distribution.head(3).index

# filtering the dataset for these top manufacturers
top_makes_data = df[df['Make'].isin(top_3_makes)]

# analyzing the popularity of EV models within these top manufacturers
ev_model_distribution_top_makes = top_makes_data.groupby(['Make', 'Model']).size().sort_values(ascending=False).reset_index(name='Number of Vehicles')

# visualizing the top 10 models across these manufacturers for clarity
top_models = ev_model_distribution_top_makes.head(10)

plt.figure(figsize=(12, 8))
sns.barplot(x='Number of Vehicles', y='Model', hue='Make', data=top_models, palette="viridis")
plt.title('Top Models in Top 3 Makes by EV Registrations')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Model')
plt.legend(title='Make', loc='center right')
plt.tight_layout()
plt.show()
```

```python
# Create a new figure with a width of 12 inches and height of 6 inches
plt.figure(figsize=(12, 6))
# Plot a histogram of the 'Electric Range' column with 30 bins and a KDE (smoothed curve)
sns.histplot(
    df['Electric Range'],  # Data to plot (miles range for EVs)
    bins=30,                    # Number of bins (controls the granularity of the histogram)
    kde=True,                   # Adds a KDE (Kernel Density Estimate) line to show distribution shape
    color='royalblue'           # Sets the color of the bars
)
# Set the title of the plot
plt.title('Distribution of Electric Vehicle Ranges')
plt.xlabel('Electric Range (miles)')
plt.ylabel('Number of Vehicles')
# Add a vertical dashed line to indicate the mean electric range in the dataset
plt.axvline(
    df['Electric Range'].mean(),  # Calculate the mean of the Electric Range
    color='red',                       # Color of the line
    linestyle='--',                    # Dashed line
    label=f'Mean Range: {df["Electric Range"].mean():.2f} miles'  # Add label with mean value
)

plt.legend()
plt.show()
```

```python
# Group the data by 'Model Year' and calculate the average 'Electric Range' for each year
average_range_by_year = df.groupby('Model Year')['Electric Range'].mean().reset_index()
# Create a new figure for the line plot with width=12 inches and height=6 inches
plt.figure(figsize=(8, 4))
# Plot a line chart to show average electric range over time
sns.lineplot(
    x='Model Year',                      # x-axis: years
    y='Electric Range',                  # y-axis: average range
    data=average_range_by_year,          # data source
    marker='o',                          # place a marker at each data point
    color='green'                        # line and marker color
)
plt.title('Average Electric Range by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Average Electric Range (miles)')
plt.grid(True)
plt.show()
```

```python
# Group the data by 'Make' and 'Model', calculate the average 'Electric Range' for each model
average_range_by_model = top_makes_data.groupby(['Make', 'Model'])['Electric Range'].mean().sort_values(ascending=False).reset_index()
# Select the top 10 models with the highest average electric range
top_range_models = average_range_by_model.head(10)
# Create a new figure for the plot (12 inches wide, 8 inches tall)
plt.figure(figsize=(12, 8))
# Create a horizontal bar plot to show average range by model, color-coded by make
barplot = sns.barplot(
    x='Electric Range',     # x-axis: average electric range
    y='Model',              # y-axis: vehicle model names
    hue='Make',             # hue: color by manufacturer (Make)
    data=top_range_models,  # dataset to use
    palette="cool"          # use the 'cool' color palette
)
plt.title('Top 10 Models by Average Electric Range in Top Makes')
plt.xlabel('Average Electric Range (miles)')
plt.ylabel('Model')
plt.legend(title='Make', loc='center right')
plt.show()
```

```python
# calculate the number of EVs registered each year
ev_registration_counts = df['Model Year'].value_counts().sort_index()
ev_registration_counts
```

```python
from scipy.optimize import curve_fit
import numpy as np

# filter the dataset to include years with complete data, assuming 2023 is the last complete year
filtered_years = ev_registration_counts[ev_registration_counts.index <= 2023]

# define a function for exponential growth to fit the data
def exp_growth(x, a, b):
    return a * np.exp(b * x)

# prepare the data for curve fitting
x_data = filtered_years.index - filtered_years.index.min()
y_data = filtered_years.values

# fit the data to the exponential growth function
params, covariance = curve_fit(exp_growth, x_data, y_data)

# use the fitted function to forecast the number of EVs for 2024 and the next five years
forecast_years = np.arange(2024, 2024 + 6) - filtered_years.index.min()
forecasted_values = exp_growth(forecast_years, *params)

# create a dictionary to display the forecasted values for easier interpretation
forecasted_evs = dict(zip(forecast_years + filtered_years.index.min(), forecasted_values))

forecasted_evs
```

```python
# prepare data for plotting
years = np.arange(filtered_years.index.min(), 2029 + 1)
actual_years = filtered_years.index
forecast_years_full = np.arange(2024, 2029 + 1)

# actual and forecasted values
actual_values = filtered_years.values
forecasted_values_full = [forecasted_evs[year] for year in forecast_years_full]

plt.figure(figsize=(12, 8))
plt.plot(actual_years, actual_values, 'bo-', label='Actual Registrations')
plt.plot(forecast_years_full, forecasted_values_full, 'ro--', label='Forecasted Registrations')

plt.title('Current & Estimated EV Market')
plt.xlabel('Year')
plt.ylabel('Number of EV Registrations')
plt.legend()
plt.grid(True)

plt.show()
```
